```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
df = pd.read_csv("RELIANCE.csv" ,parse_dates=["Date"])
df
```

```python
df.set_index("Date",inplace=True)
df
```

```python
df.rename(columns={
    "Open Price":"Open",
    "Close Price":"Close",
    "Low Price":"Low",
    "High Price":"High",
    "Total Traded Quantity":"Volume"
},inplace=True)
df
```

```python
s = mpf.make_mpf_style(base_mpf_style='yahoo',y_on_right=False,gridstyle=':')
mpf.plot(df.head(100),mav=(9,26),type='candlestick', style=s, volume=True )
```
