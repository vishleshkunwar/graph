```python
import pandas as pd
```

```python
rng = pd.date_range(start="06/30/2021",end="07/02/2021", freq='b')
rng = rng.strftime('%d%m%y')
rng.to_list()
```

```python
for i in rng:
    date_v = pd.to_datetime(i,format="%d%m%y")
    date_v = date_v.strftime('%d-%m-%Y')
    file_name = "bsedata/EQ"+ i +".CSV"
    try:
        df = pd.read_csv(file_name)
        df['Date'] = date_v
        columns = ['Date' , 'SC_CODE', 'SC_NAME', 'OPEN', 'HIGH', 'LOW',
               'CLOSE', 'LAST', 'PREVCLOSE', 'NO_TRADES', 'NO_OF_SHRS', 'NET_TURNOV']
        df = df[columns]
        for i in range(0,df.size):
            file_name =str(df.loc[i]['SC_CODE'])
            name_company = df.loc[i]['SC_NAME']
            data = pd.DataFrame(df.loc[i])
            data = pd.DataFrame.transpose(data)
            try:
                company_df = pd.read_csv("bsedatacompany/"+file_name + ".csv")
                data = data.append(company_df)
                data.to_csv("bsedatacompany/"+file_name + ".csv",index=False)
            except:
                data.to_csv("bsedatacompany/"+file_name + ".csv",index=False)
                print(name_company)    
    except:
        print(" ")
```
