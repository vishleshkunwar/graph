```python
#create table 'stockdata'

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="vishal",
database='company')

mycursor = mydb.cursor()

mycursor.execute("CREATE TABLE stockdata(user_id int auto_increment primary key,Date DATE,Symbol VARCHAR(20),Prev_close double,open double,high double,low double,last_price double,close double,volume double,no_of_trades int,stock_exchange varchar(5))")
```

```python
company_csv = pd.read_csv('nselist.csv')
size=company_csv.size
for i in range(1000,size):
    try:
        company=company_csv.SYMBOL[i]
        df=pd.read_csv(r'C:/Users/91766/Desktop/learn/nsedata/' + company+'.csv',index_col="Date",parse_dates=True)
        df=df.reset_index()
        df=df.rename(columns={'Open Price':'open','High Price':'high','Low Price':'low','Close Price':'close','Total Traded Quantity':'volume'})
        df=df[['Date', 'Symbol','Prev Close', 'open', 'high', 'low','Last Price', 'close','volume','No. of Trades']]

        df['stock_exchange']='nse'
        df['No. of Trades'].replace({"-":0},inplace=True)
        
        sql = "INSERT INTO stockdata (Date,Symbol,Prev_close,open,high,low,last_price,close,volume,no_of_trades,stock_exchange) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        val = df.values.tolist()
        mycursor.executemany(sql,val)

        mydb.commit()

        print(mycursor.rowcount, "was inserted.")
    except:
        print(company,'was fail to insert in database')
```
